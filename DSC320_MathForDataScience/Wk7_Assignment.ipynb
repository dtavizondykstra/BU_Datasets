# **Working with Matrices**

Let **A**, **B**, and **C** be the matrices given below.

$$A = \begin{bmatrix} 1 & -2 & 3 & 7 \\ 2 & 1 & 1 & 4 \\ -3 & 2 & -2 & 10 \end{bmatrix} \quad
B = \begin{bmatrix} 0 & 1 & -3 & -2 \\ 10 & -1 & 2 & -3 \\ 5 & 1 & -1 & 4 \end{bmatrix} \quad
C = \begin{bmatrix} 4 & 0 & -2 & 3 \\ 3 & 6 & 9 & 7 \\ 2 & 2 & 5 & 1 \\ 9 & 4 & 6 & -2 \end{bmatrix}$$

## **Problem 1: Define Matrices A, B, and C**

Enter each of the matrices A, B, and C as a NumPy matrix.

```python
# ---------------------------------------------------------
# PROBLEM 1: DEFINE MATRICES
# ---------------------------------------------------------
# Enter matrices A, B, and C as NumPy arrays.

A = np.array([
    [ 1, -2,  3,  7],
    [ 2,  1,  1,  4],
    [-3,  2, -2, 10]
])

B = np.array([
    [ 0,  1, -3, -2],
    [10, -1,  2, -3],
    [ 5,  1, -1,  4]
])

C = np.array([
    [4, 0, -2,  3],
    [3, 6,  9,  7],
    [2, 2,  5,  1],
    [9, 4,  6, -2]
])

print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)
print("\nMatrix C:")
print(C)
```

## **Problem 2: Matrix Dimensions**

What are the dimensions of each of the matrices A, B, and C? Display these using Python code.

```python
# ---------------------------------------------------------
# PROBLEM 2: MATRIX DIMENSIONS
# ---------------------------------------------------------
# .shape returns (rows, columns) for each matrix.

print(f"Dimensions of A: {A.shape}")
print(f"Dimensions of B: {B.shape}")
print(f"Dimensions of C: {C.shape}")
```

**Answer**: Matrix A is 3×4, Matrix B is 3×4, and Matrix C is 4×4.

## **Problem 3: Matrix Operations**

Determine if each of the following matrix operations is defined. If it is defined, calculate the result using Python. If it is not defined, explain why it is not defined.

### **Problem 3A: A + B**

```python
# ---------------------------------------------------------
# PROBLEM 3A: A + B
# ---------------------------------------------------------
# Matrix addition requires both matrices to have the same dimensions.
# A is 3x4 and B is 3x4 — dimensions match, so addition is defined.

result_AplusB = A + B

print("A + B =")
print(result_AplusB)
```

**Answer**: A + B is **defined**. Both A and B are 3×4 matrices, so element-wise addition is valid.

### **Problem 3B: AB**

```python
# ---------------------------------------------------------
# PROBLEM 3B: AB (matrix multiplication)
# ---------------------------------------------------------
# Matrix multiplication A @ B requires the number of columns in A
# to equal the number of rows in B.
# A is 3x4 (4 columns) and B is 3x4 (3 rows) — these do not match.

print(f"A has shape: {A.shape} -> {A.shape[1]} columns")
print(f"B has shape: {B.shape} -> {B.shape[0]} rows")
print(f"\nAB is defined: {A.shape[1] == B.shape[0]}")
```

**Answer**: AB is **not defined**. Matrix multiplication requires the number of columns in A (4) to equal the number of rows in B (3). Since 4 ≠ 3, this operation is undefined.

### **Problem 3C: AC**

```python
# ---------------------------------------------------------
# PROBLEM 3C: AC (matrix multiplication)
# ---------------------------------------------------------
# A is 3x4 (4 columns) and C is 4x4 (4 rows) — dimensions match.
# The result will be a 3x4 matrix.

print(f"A has shape: {A.shape} -> {A.shape[1]} columns")
print(f"C has shape: {C.shape} -> {C.shape[0]} rows")
print(f"\nAC is defined: {A.shape[1] == C.shape[0]}")
print()

result_AC = A @ C

print("AC =")
print(result_AC)
```

**Answer**: AC is **defined**. A is 3×4 and C is 4×4 — the inner dimensions match (4 = 4), so the multiplication is valid and produces a 3×4 matrix.

### **Problem 3D: C$^T$**

```python
# ---------------------------------------------------------
# PROBLEM 3D: TRANSPOSE OF C
# ---------------------------------------------------------
# The transpose of a matrix flips rows and columns.
# C is 4x4, so C^T is also 4x4.

result_CT = C.T

print("C^T =")
print(result_CT)
```

**Answer**: C$^T$ is **defined**. The transpose of any matrix is always defined. Since C is 4×4, its transpose C$^T$ is also 4×4.

## **Problem 4: Transpose Property — $(A + B)^T = A^T + B^T$**

Illustrate the following property of matrix transpositions using these matrices: $(A + B)^T = A^T + B^T$.

```python
# ---------------------------------------------------------
# PROBLEM 4: TRANSPOSE PROPERTY (A + B)^T = A^T + B^T
# ---------------------------------------------------------
# Calculate both sides and verify they are equal.

left_side  = (A + B).T
right_side = A.T + B.T

print("Left side  — (A + B)^T =")
print(left_side)

print("\nRight side — A^T + B^T =")
print(right_side)

print(f"\nAre both sides equal? {np.allclose(left_side, right_side)}")
```

**Answer**: Both sides are equal, which confirms the property $(A + B)^T = A^T + B^T$. Transposing the sum of two matrices is equivalent to summing their individual transposes.

## **Problem 5: Transpose Property — $(AC)^T = C^T A^T$**

Illustrate the following property of matrix transposition using these matrices: $(AC)^T = C^T A^T$.

```python
# ---------------------------------------------------------
# PROBLEM 5: TRANSPOSE PROPERTY (AC)^T = C^T A^T
# ---------------------------------------------------------
# Calculate both sides and verify they are equal.

left_side  = (A @ C).T
right_side = C.T @ A.T

print("Left side  — (AC)^T =")
print(left_side)

print("\nRight side — C^T A^T =")
print(right_side)

print(f"\nAre both sides equal? {np.allclose(left_side, right_side)}")
```

**Answer**: Both sides are equal, confirming the property $(AC)^T = C^T A^T$. Note that the order of the matrices is reversed on the right-hand side — this reversal is required for the dimensions to align correctly in the multiplication.

## **Problem 6: Inverse of C**

Find $C^{-1}$ using Python code.

```python
# ---------------------------------------------------------
# PROBLEM 6: INVERSE OF C
# ---------------------------------------------------------
# np.linalg.inv() computes the inverse of a square matrix.
# A matrix is invertible only if its determinant is non-zero.

det_C = np.linalg.det(C)
print(f"Determinant of C: {det_C:.6f}")
print(f"C is invertible: {not np.isclose(det_C, 0)}")
print()

C_inv = np.linalg.inv(C)

print("C^-1 =")
print(np.round(C_inv, 6))
```

```python
# ---------------------------------------------------------
# VERIFY: C * C^-1 should equal the identity matrix
# ---------------------------------------------------------

identity_check = np.round(C @ C_inv, 6)

print("C @ C^-1 (should be identity matrix):")
print(identity_check)
```

**Answer**: The inverse $C^{-1}$ exists because the determinant of C is non-zero. We can verify the result by confirming that $C \cdot C^{-1} = I$, where $I$ is the identity matrix.

# **Working with Images**

For this exercise, we will be working with the image file `week7_image.jpg`.

## **Problem 1: Display the Image**

Use the following code to display the image.

```python
# ---------------------------------------------------------
# PROBLEM 1: DISPLAY THE IMAGE
# ---------------------------------------------------------
# OpenCV imread() only reads local file paths, so we fetch
# the image bytes from GitHub and decode them into a NumPy array.

url = 'https://raw.githubusercontent.com/dtavizondykstra/BU_Datasets/refs/heads/main/DSC320_MathForDataScience/week7_image.jpg'
response = requests.get(url)

# decode bytes into NumPy array (OpenCV format is BGR by default)
image_bgr = cv2.imdecode(np.frombuffer(response.content, np.uint8), cv2.IMREAD_COLOR)

# convert to RGB for display with matplotlib
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

# display the image
plt.imshow(image_rgb)
plt.axis("off")
plt.show()
```

## **Problem 2: Image Resolution**

Find the resolution of this image. **Hint**: OpenCV imports the image as a NumPy array, where each pixel corresponds to an entry in the array.

```python
# ---------------------------------------------------------
# PROBLEM 2: IMAGE RESOLUTION
# ---------------------------------------------------------
# The shape of the NumPy array gives us height, width, and channels.
# Resolution is expressed as width x height.

print(f"Array shape: {image_rgb.shape}")
print(f"\nImage resolution: {image_rgb.shape[1]} x {image_rgb.shape[0]} pixels")
print(f"  Width  : {image_rgb.shape[1]} pixels")
print(f"  Height : {image_rgb.shape[0]} pixels")
```

**Answer**: The image resolution is **600 × 453 pixels** (width × height). This is read directly from the NumPy array shape, where the first dimension is height (rows) and the second is width (columns).

## **Problem 3: The Third Dimension**

In checking the array dimensions, you should see that three numbers are displayed. What is this third number, and why is it there?

```python
# ---------------------------------------------------------
# PROBLEM 3: THE THIRD DIMENSION
# ---------------------------------------------------------
# Display the full array shape and identify each dimension.

print(f"Full array shape: {image_rgb.shape}")
print()
print(f"  Dimension 1 (height)   : {image_rgb.shape[0]} rows of pixels")
print(f"  Dimension 2 (width)    : {image_rgb.shape[1]} columns of pixels")
print(f"  Dimension 3 (channels) : {image_rgb.shape[2]} color channels (R, G, B)")
```

**Answer**: The third number is **3**, representing the three color channels — **Red**, **Green**, and **Blue** (RGB). 

Each pixel in a color image is not a single value but a combination of three intensity values, one per channel, each ranging from 0 to 255. Together these three values define the exact color of that pixel. A grayscale image would only have one channel, so its array shape would be (height, width) with no third dimension.
