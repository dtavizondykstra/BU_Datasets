# **Calculating Limits Numerically**

Let $f(x) = 2x^3 - 4x + 1$ and let $g(x) = \frac{e^x - 1}{x}$

### **Problem A**
(a) Use Python code to evaluate the function $f$ at appropriate values to calculate $\lim_{x \to 3} f(x)$.

**Hint:** Evaluate $f$ at values closer and closer to 3 and see what value the outputs get close to.

```python
def f(x):
    """
    Polynomial function f(x) = 2x^3 - 4x + 1.

    Parameters
    ----------
    x : float or array_like
        The value(s) at which to evaluate the function.

    Returns
    -------
    float or array_like
        The function value(s).
    """
    return 2 * x**3 - 4 * x + 1

# Evaluate f(x) as x approaches 3 from both sides
vals_near_3 = [2.9, 2.99, 2.999, 3.001, 3.01, 3.1]
print(f"Evaluating f(x) near x=3:")
for v in vals_near_3:
    print(f"f({v:.3f}) = {f(v):.5f}")

print("\nEstimated limit as x → 3:", f(3))
```

### **Problem B**
Use Python code to evaluate the function $g$ at appropriate values to calculate $\lim_{x \to 0} g(x)$

```python
def g(x):
    """
    Function g(x) = (e^x - 1) / x.

    Parameters
    ----------
    x : float or array_like
        The value(s) at which to evaluate the function.

    Returns
    -------
    float or array_like
        The function value(s).
    """
    return (np.exp(x) - 1) / x

# Evaluate g(x) as x approaches 0
vals_near_0 = [-0.1, -0.01, -0.001, 0.001, 0.01, 0.1]
print(f"Evaluating g(x) near x=0:")
for v in vals_near_0:
    print(f"g({v:.3f}) = {g(v):.5f}")

print("\nEstimated limit as x → 0:", 1.0)
print("The function approaches 1 as x approaches 0 from both sides.")
```

### **Problem C**
(c) Could the limits in parts (a) and (b) be calculated by plugging in the value where the limit is taken
into the function? Explain. What does this mean in terms of continuity?

**Answer**:
* For part (a), the limit can be calculated by directly plugging in $x = 3$ because $f(x)$ is a polynomial function, which is continuous everywhere. So $\lim_{x \to 3} f(x) = f(3)$.
* For part (b), we cannot plug in $x = 0$ directly into $g(x)$ because it would
result in division by zero, which is undefined. However, by evaluating
values close to 0, we can see that the function approaches a specific value.
* This means that $g(x)$ has a limit at $x = 0$ even though it is not defined
at that exact point. This is an example of a removable discontinuity (Abramson, 2021).

# **Average Rate of Change**

The formula for the average rate of change of a function $f$ between two values $a$ and $b$ is given by:
$$\text{Average Rate of Change} = \frac{f(b) - f(a)}{b - a}$$

Create a Python function that takes in a mathematical function (e.g., $f(x) = 3x^2$) and two numbers $a$ and $b$, and returns the average rate of change of the function between $a$ and $b$.

```python
def average_rate_of_change(func, a, b):
    """
    Calculate the average rate of change of a function over the interval [a, b].

    Parameters
    ----------
    func : callable
        The mathematical function to evaluate.
    a : float
        The starting point of the interval.
    b : float
        The ending point of the interval.

    Returns
    -------
    float
        The average rate of change: (f(b) - f(a)) / (b - a).
    """
    # Ensure we don't divide by zero if a equals b
    if a == b:
        raise ValueError("a and b must be different values")
    return (func(b) - func(a)) / (b - a)

# Example usage with f(x) = 3x^2
example_func = lambda x: 3 * x**2
print(f"Average rate of change of 3x^2 from 1 to 2: {average_rate_of_change(example_func, 1, 2)}")
```

# **Average Rate of Change to Instantaneous Rate of Change**
A baseball is dropped from a tall cliff. Neglecting air resistance, the distance traveled by the baseball in meters after $t$ seconds is given by the function:
$$f(t) = 4.9t^2$$


### **Problem A**

(a) Find the average speed (average rate of change of distance) of the baseball between 5 and 6 seconds.

```python
def baseball_dist(t):
    """
    Calculate distance traveled by a falling baseball.

    Parameters
    ----------
    t : float
        Time in seconds.

    Returns
    -------
    float
        Distance in meters.
    """
    return 4.9 * t**2

# Problem A: Average speed between 5 and 6 seconds
speed_5_6 = average_rate_of_change(baseball_dist, 5, 6)
print(f"Average speed (5 to 6s): {speed_5_6:.2f} m/s")
```

### **Problem B**
(b) Find the average speed (average rate of change of distance) of the baseball between 5 and 5.5 seconds.

```python
# Problem B: Average speed between 5 and 5.5 seconds
speed_5_55 = average_rate_of_change(baseball_dist, 5, 5.5)
print(f"Average speed (5 to 5.5s): {speed_5_55:.2f} m/s")
```

### **Problem C**
(c) Find the average speed (average rate of change of distance) of the baseball between 5 and 5.1 seconds.




```python
# Problem C: Average speed between 5 and 5.1 seconds
speed_5_51 = average_rate_of_change(baseball_dist, 5, 5.1)
print(f"Average speed (5 to 5.1s): {speed_5_51:.2f} m/s")
```

### **Problem D**
(d) What is the instantaneous speed of the baseball at $t = 5$ seconds?

**Hint:** You might want to do a few more calculations similar to those done in parts (a) - (c).

```python
# Problem D: Approximating instantaneous speed at t=5
for delta in [0.1, 0.01, 0.001, 0.0001]:
    avg_speed = average_rate_of_change(baseball_dist, 5, 5 + delta)
    print(f"Interval [5, 5 + {delta}]: {avg_speed:.4f} m/s")

print("""
  As the interval gets smaller, the average speed approaches 49.0 m/s,
  which represents the instantaneous speed at t = 5 seconds.
""")
```

### **Problem E**
(e) Find the derivative of $f$.

**Hint:** The derivative of a quadratic function $f(x) = ax^2 + bx + c$ is given by $f'(x) = 2ax + b$.

```python
def baseball_derivative(t):
    """
    Calculates the derivative of f(t) = 4.9t^2.

    Using the power rule: f'(t) = 2 * 4.9 * t = 9.8t.
    """
    return 9.8 * t

print(f"The derivative function is f'(t) = 9.8t")
```

### **Problem F**
(f) Evaluate the derivative of $f$ at $t = 5$. How does this value compare to the value found in part (d)?
Explain what is happening

```python
# Problem F: Evaluate derivative at t=5
inst_speed = baseball_derivative(5)
print(f"Instantaneous speed at t=5: {inst_speed} m/s")

print("""
  This matches the limit found in part (d).
  This shows that the derivative represents the instantaneous rate of change,
  which is the limit of the average rate of change as the interval approaches zero.
""")
```

# **Calculating and Interpreting Partial Derivatives**

A multivariate linear model is trained to predict the selling price of a particular used car model. The linear model predicts the selling price ($P$) in US dollars, using its current condition ($C$) on a scale of 1–10 and the age of the car in years ($Y$).

The model is given by:
$$P = 16,000 + 2,400C - 1,800Y$$

### **Problem A**
(a) What is this model’s predicted selling price of a 5-year old car with a condition rating 8?

```python
def car_price(C, Y):
    """
    Predicts car price based on condition and age.
    P = 16000 + 2400C - 1800Y
    """
    return 16000 + 2400*C - 1800*Y

# Problem A: 5-year old car with condition 8
price = car_price(8, 5)
print(f"Predicted price: ${price:,.2f}")
```

### **Problem B**
(b) Find $\frac{\partial P}{\partial C}$ and interpret this value.

```python
# Problem B: Partial derivative with respect to C
# dP/dC = 2400
print("dP/dC = 2400")
print("Interpretation: For every 1-unit increase in condition, the price increases by $2,400.")
```

### **Problem C**
(c) Find $\frac{\partial P}{\partial Y}$ and interpret this value.

```python
# Problem C: Partial derivative with respect to Y
# dP/dY = -1800
print("dP/dY = -1800")
print("Interpretation: For every 1-year increase in age, the price decreases by $1,800.")
```
