# **Data Scaling and Working with Dataframes**
When training models on data, often times it is useful to scale input features, so they all have a similar range of values. There are multiple ways to apply this scaling, and you will build functions to perform some of these. You will then apply this on some data.

## **Problem 1: Data Normalization**

A common method to scale a dataset feature is to normalize it, placing all values on a scale between 0 and 1. Given a set of values $X_1, X_2, \ldots, X_n$, each normalized value is calculated as:

$$\text{normalized } X_i = \frac{X_i - \min\{X_1, X_2, \ldots, X_n\}}{\max\{X_1, X_2, \ldots, X_n\} - \min\{X_1, X_2, \ldots, X_n\}}$$

Build a Python function that takes in a vector (array) and normalizes it.

```python
# ---------------------------------------------------------
# DEFINE NORMALIZATION FUNCTION
# ---------------------------------------------------------
# Scales all values in a vector to the range [0, 1]
# using the min-max formula.

def normalize(x):
    """
    Normalize a vector to the range [0, 1].

    Parameters:
        x (array-like): input vector of numeric values

    Returns:
        np.ndarray: normalized vector
    """

    x = np.array(x, dtype=float)

    x_min = np.min(x)
    x_max = np.max(x)

    return (x - x_min) / (x_max - x_min)
```

## **Problem 2: Data Standardization**

Another common method is standardization, which calculates the z-score of each value. Given a set of values $X_1, X_2, \ldots, X_n$ with sample mean $\bar{X}$ and sample standard deviation $s_X$, each standardized value is calculated as:

$$\text{standardized } X_i = \frac{X_i - \bar{X}}{s_X}$$

Build a Python function that takes in a vector (array) and standardizes it.

```python
# ---------------------------------------------------------
# DEFINE STANDARDIZATION FUNCTION
# ---------------------------------------------------------
# Scales values by subtracting the mean and dividing by
# the sample standard deviation (z-score transformation).

def standardize(x):
    """
    Standardize a vector to have mean 0 and standard deviation 1.

    Parameters:
        x (array-like): input vector of numeric values

    Returns:
        np.ndarray: standardized vector (z-scores)
    """

    x = np.array(x, dtype=float)

    x_mean = np.mean(x)
    x_std  = np.std(x, ddof=1)  # ddof=1 for sample standard deviation

    return (x - x_mean) / x_std
```

### **Problem 3C**
Create a new feature by dividing total bedrooms by the number of households. What does this new feature represent?

```python
# ---------------------------------------------------------
# PROBLEM 3C: CREATE NEW FEATURE
# ---------------------------------------------------------
# Dividing total bedrooms by total households gives the
# average number of bedrooms per household on the block.
#
# Note: total_bedrooms has 207 null values. We impute these
# with the column median before creating the feature so that
# all rows are retained and standardization works correctly.

df['total_bedrooms'] = df['total_bedrooms'].fillna(df['total_bedrooms'].median())
df['bedrooms_per_household'] = df['total_bedrooms'] / df['households']

print("New feature (first 10 values):")
print(df['bedrooms_per_household'].head(10))
```

**Answer**: The new feature `bedrooms_per_household` represents the average number of bedrooms per household on the block. This is more informative than total bedrooms alone, since it accounts for the number of households and gives a per-unit measure of bedroom density.

### **Problem 3D**

### **Problem 3E**
Apply data standardization to the features in the dataframe from 3D.

```python
# ---------------------------------------------------------
# PROBLEM 3E: STANDARDIZE FEATURE DATAFRAME
# ---------------------------------------------------------
# Apply the standardize() function to each column.
# Result is a new dataframe with z-scored features.

new_df_standardized = new_df.apply(standardize)

print("Standardized feature dataframe (first 10 rows):")
new_df_standardized.head(10)
```

```python
# ---------------------------------------------------------
# VERIFY STANDARDIZATION
# ---------------------------------------------------------
# Each column should have mean ≈ 0 and std ≈ 1.

print("Column means (should be ≈ 0):")
print(new_df_standardized.mean().round(6))

print("\nColumn std devs (should be ≈ 1):")
print(new_df_standardized.std().round(6))
```
